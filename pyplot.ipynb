2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

```python
def generate_random(seed):
    ''' This function is a simple implementation
        of the linear congruential generator (LCG) algorithm. It takes 3
        paramaters: a, c, and m.
        
        a: The multiplier. It is a large prime number that determines 
        the period of the generator. It should be chosen in such a way 
        that it is coprime to the modulus m for better randomness. 
        In this case, the value a = 1664525 is a commonly used multiplier.
        
        c: The increment. It is a constant added to the product of the 
        previous random number and the multiplier. It helps introduce 
        additional randomness and avoid repetitions. 
        The value c = 1013904223 is another commonly used constant.
        
        m: The modulus. It is a large prime number that defines 
        the range of the generated random numbers. In this case, 
        m = 2**32 is used to generate 32-bit unsigned integers.'''
    
    a = 1664525
    c = 1013904223
    m = 2**32
    num = seed

    while True:
        num = (a * num + c) % m
        yield num

generator = generate_random(12345)

# Generate 5 random numbers
random_numbers = []
for _ in range(5):
    random_num = next(generator)
    random_numbers.append(random_num)
    print(random_num)
```

```python
 x_values = [random_numbers.index(x) for x in random_numbers]
```

```python
plt.scatter(x_values, random_numbers)
plt.xticks(range(5))
plt.ticklabel_format(style='plain', axis='y');
plt.title("Scatter Plot Randomly Generated Numbers")
plt.xlabel("Numbers")
plt.ylabel("Value")
plt.show();
```

```python
plt.plot(x_values,random_numbers)
plt.xticks(range(5))
#plt.ticklabel_format(style='plain', axis='y')
plt.title("Randomly Generated Numbers Line Plot")
plt.xlabel("Numbers")
plt.ylabel("Value")
plt.show();
```
